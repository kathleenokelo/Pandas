```python
#Step 1. Import the necessary libraries

import pandas as pd
from pandas.tseries.offsets import BMonthBegin , BMonthEnd
```

```python
#Step 2. Import the dataset from this address
#Step 3. Assign it to a variable apple

apple = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv')
apple
```

```python
#Step 4. Check out the type of the columns

apple.dtypes
```

```python
#Step 5. Transform the Date column as a datetime type

apple['Date'] = pd.to_datetime(apple['Date'])
apple.dtypes
```

```python
#Step 6. Set the date as the index

idxapple = apple.set_index(['Date'])
idxapple
```

```python
#Step 7. Is there any duplicate dates?
duplicate_dates = apple[apple['Date'].duplicated(keep=False)]
duplicate_dates
```

```python
#Step 8. Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

sorted_apple = apple.sort_values(by='Date')
sorted_apple
```

```python
#Step 9. Get the last business day of each month

startdate = '1980-12-12'
enddate = '2014-07-08'

date_range = pd.date_range(start=startdate,end=enddate,freq='MS')

last_business_days = date_range + BMonthEnd(0)

print(last_business_days)
```

```python
#Step 10. What is the difference in days between the first day and the oldest

import pandas as pd
days_difference =  pd.Timestamp('2014-07-08') - pd.Timestamp('1980-12-12')
print('Difference between the days:',days_difference)
```

```python
#Step 11. How many months in the data we have?

months = apple['Date'].dt.month
no_of_months = months.nunique()

print('Number of months:',no_of_months)
```

```python
#Step 1. Import the necessary libraries

import pandas as pd
```

```python
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called iris

iris = pd.read_table('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',sep=',')
iris
```

```python
#Step 4. Create columns for the dataset
 #1. sepal_length (in cm)
 #2. sepal_width (in cm)
 #3. petal_length (in cm)
 #4. petal_width (in cm)
 #5. class

iris.columns=(['sepal_length(in cm)','sepal_width(in cm)','petal_length(in cm)','petal-width(in cm)','class'])
iris
```

```python
#Step 5. Is there any missing value in the dataframe?

iris.isnull().any()
```

```python
#Step 6. Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

import numpy as np

iris.loc[10:30,'petal_length(in cm)'] = np.nan
print(iris.loc[10:30,'petal_length(in cm)'] )


```

```python
#Step 7. Good, now lets substitute the NaN values to 1.0

iris.loc[10:30,'petal_length(in cm)'] = iris.loc[10:30,'petal_length(in cm)'].fillna(value=1.0)
print(iris.loc[10:30,'petal_length(in cm)'] )
```

```python
#Step 8. Now let's delete the column class
iris.drop(columns='class',inplace=True)
iris.head()
```

```python

#Step 9. Set the first 3 rows as NaN
iris.iloc[0:3] = np.nan
iris.head()
```

```python
#Step 10. Delete the rows that have NaN
iris_clean = iris.dropna()
iris_clean.head()
```

```python
#Step 11. Reset the index so it begins with 0 again

iris_clean.reset_index(drop=True,inplace=True)
iris_clean.head()
```
