```python
#Step 1. Import the necessary libraries
import pandas as pd
```

```python
#Step 2. Import the dataset from this address.
#Step 3. Assign it to a variable called baby_names.

baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
baby_names
```

```python
#Step 4. See the first 10 entries

baby_names.head(10)
```

```python
#Step 5. Delete the column 'Unnamed: 0' and 'Id'

del baby_names['Unnamed: 0']
del baby_names['Id']

baby_names
```

```python
#Step 6. Is there more male or female names in the dataset?

female = baby_names['Gender'] == 'F'
number_of_females = female.sum()
print('Number of females:',number_of_females)

male = baby_names['Gender'] == 'M'
number_of_males = male.sum()
print('Number of males:' ,number_of_males)

if number_of_females > number_of_males:
    print(f'There are {number_of_females - number_of_males} more females than males ')
else:
    print(f'There are {number_of_males - number_of_females} less females than males')
```

```python
#Step 7. Group the dataset by name and assign to names

names = baby_names.groupby(by='Name')['Count'].sum()
names
```

```python
#Step 8. How many different names exist in the dataset?

no_of_names = baby_names['Name'].nunique()
no_of_names
```

```python
#Step 9. What is the name with most occurrences?

most_common_name = names.idxmax()
count = names.max()

print(f'The most common name is {most_common_name} with {count} occurences')
```

```python
#Step 10. How many different names have the least occurrences

least_common_name = names.idxmin()
count = names.min()

print(f'The least common name is {least_common_name} is {count} occurences')
```

```python
#Step 11. What is the median name occurrence?

median_name = names.median()
print(f'The median occurences is {median_name}')
```

```python
#Step 12. What is the standard deviation of names?

std_names = names.std()
print(f'The standard deviation of name occurences is {std_names}')
```

```python
#Step 13. Get a summary with the mean, min, max, std and quartiles.

baby_names.describe()
```
